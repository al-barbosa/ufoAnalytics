```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("UFOs_coord.csv", encoding = "ISO-8859-1")
df.head()
```

```python
df = df.rename(columns={"Date / Time": "date_time", "Country": "country", "City": "city", "State": "state", "Shape": "shape", "Summary": "summary"})
df.head()
```

```python
def date_convert(date_string):
    if(len(date_string) > 15):
        return pd.to_datetime(date_string, format="%Y-%m-%d %H.%M.%S")
    elif(len(date_string) > 10):
        return pd.to_datetime(date_string, format="%m/%d/%y %H:%M")
    else:
        pd.to_datetime(date_string, format="%m/%d/%y")
```

```python
df["date_time"] = df["date_time"].apply(lambda x: date_convert(x))
df.head()
```

```python
df = df[df["date_time"].notnull()]
```

```python
df["year"] = df["date_time"].apply(lambda x: int(x.year))
df["month"] = df["date_time"].apply(lambda x: int(x.month))
df.head()
```

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_counts = df['month'].value_counts().sort_index()

plt.plot(month_counts.index, month_counts.values, marker='o', linestyle='-', color='blue')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.title('Frequency in Months')
plt.xticks(month_counts.index, [month_order[i-1] for i in month_counts.index])
plt.ylim(0)
plt.grid(True)
plt.show()
```

```python
map_center = [df['lat'].mean(), df['lng'].mean()]
map_zoom = 4  # Adjust the zoom level as needed
m = folium.Map(location=map_center, zoom_start=map_zoom)

# Add markers for each latitude and longitude point
for index, row in df.iterrows():
    lat, lon = row['lat'], row['lng']
    folium.Marker([lat, lon]).add_to(m)
    
m
```

```python
df["summary"] = df["summary"].str.lower()
df["summary"]
```

```python
stop_words = set(stopwords.words("english"))
df["summary_no_stop"] = df["summary"].apply(lambda x: " ".join([word for word in x.split() if word not in stop_words]))
df["summary_no_stop"]
```

```python
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(df["summary_no_stop"])

word_frequencies = pd.DataFrame(bag_of_words.toarray(), columns=vectorizer.get_feature_names()).sum().sort_values(ascending=False)
word_frequencies[:15]
```
